```python
import pandas as pd

game_metadata_path = '../data/game_metadata.csv'
# read the data and store data in DataFrame titled melbourne_data
game_metadata_data = pd.read_csv(game_metadata_path) 
# print a summary of the data in Melbourne data
game_metadata_data.describe()
```

```python
# For each column in the DataFrame, display unique values, data type, and number of missing values
for column in game_metadata_data.columns:
    print(f"Column: {column}")
    print(f"  Data type: {game_metadata_data[column].dtype}")
    print(f"  Number of missing values: {game_metadata_data[column].isnull().sum()}")
    unique_vals = game_metadata_data[column].unique()
    print(f"  Number of unique values: {len(unique_vals)}")
    # Show up to 10 unique values for preview
    print(f"  Sample unique values: {unique_vals[:10]}")
    print("-" * 40)
```
